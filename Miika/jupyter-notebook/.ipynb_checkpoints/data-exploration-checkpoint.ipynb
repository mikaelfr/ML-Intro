Task A)

```python
def f(x):
    return (1 + 𝑥) - ((𝑥**2)/2)

def sample_pairs(n=1):
    x_arr = []
    y_arr = []
    for i in range(n):
        x = np.random.uniform(low=-3.0, high=3.0)
        epsilon = np.random.normal(0, 0.4, 1)
        y = f(x) + epsilon[0]
        x_arr.append(x)
        y_arr.append(y)
    return np.array([x_arr, y_arr])

plot = sample_pairs(100)
training = sample_pairs(20)
validation = sample_pairs(20)
test = sample_pairs(1000)
```

Task B)

```python
from matplotlib.pyplot import figure
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import operator


figure(figsize=(10, 10), dpi=500)

# tr = training
x_tr = training[0][:,np.newaxis]
y_tr = training[1][:,np.newaxis]

x_val = validation[0][:,np.newaxis]
y_val = validation[1][:,np.newaxis]

x_test = test[0][:,np.newaxis]
y_test = test[1][:,np.newaxis]

x_plot = plot[0][:,np.newaxis]

mse_tr = []
mse_val = []
mse_test = []

for degree in range(0,11):
    polynomial_features= PolynomialFeatures(degree=degree)
    
    x_poly_tr = polynomial_features.fit_transform(x_tr)
    x_poly_plot = polynomial_features.fit_transform(x_plot)
    
    x_poly_val = polynomial_features.fit_transform(x_tr)
    x_poly_test = polynomial_features.fit_transform(x_test)

    model = LinearRegression()
    model.fit(x_poly_tr, y_tr)
    
    y_poly_pred_tr = model.predict(x_poly_tr)
    y_poly_pred_val = model.predict(x_poly_val)
    y_poly_pred_test = model.predict(x_poly_test)
    
    y_poly_pred_plot = model.predict(x_poly_plot)

    mse_tr.append(mean_squared_error(y_tr,y_poly_pred_tr))
    mse_val.append(mean_squared_error(y_val,y_poly_pred_val))
    mse_test.append(mean_squared_error(y_test,y_poly_pred_test))


    plt.scatter(x_tr, y_tr, s=10)
    # sort the values of x before line plot
    sort_axis = operator.itemgetter(0)
    sorted_zip = sorted(zip(x_plot ,y_poly_pred_plot), key=sort_axis)
    x, y = zip(*sorted_zip)
    plt.plot(x, y)

plt.show()
```

```python
mse_data = np.column_stack([range(0,11), mse_tr])

mses = pd.DataFrame(data=mse_data, columns=['degree', 'mse_tr'], dtype="object")

mses
```

Task C)

```python
mses['mse_val'] = mse_val
mses['mse_val'] = mses['mse_val'].astype("object")
mses['mse_test'] = mse_test
mses['mse_test'] = mses['mse_test'].astype("object")

mses
```

Task D)

```python
from sklearn.model_selection import KFold

x_comb = np.concatenate((training[0], validation[0]), axis=None)[:,np.newaxis]
y_comb = np.concatenate((training[1], validation[1]), axis=None)[:,np.newaxis]

mse_kfold = []

kf = KFold(n_splits=10)

for degree in range(0,11):
    splits_mse = []
    for test, train in kf.split(list(zip(x_comb, y_comb))):
        x_comb_tr = [x_comb[i] for i in train]
        y_comb_tr = [x_comb[i] for i in train]
        
        x_comb_test = [x_comb[i] for i in test]
        y_comb_test = [x_comb[i] for i in test]
                        
        polynomial_features= PolynomialFeatures(degree=degree)
    
        x_poly_comb_tr = polynomial_features.fit_transform(x_comb_tr)
        x_poly_comb_test = polynomial_features.fit_transform(x_comb_test)
        
        model = LinearRegression()
        model.fit(x_poly_comb_tr, y_comb_tr)
            
        y_poly_comb_pred = model.predict(x_poly_comb_test)
                
        splits_mse.append(mean_squared_error(y_comb_test, y_poly_comb_pred))
    mse_kfold.append(np.average(splits_mse))
    
mses['mse_kfold'] = mse_kfold
mses['mse_kfold'] = mses['mse_kfold'].astype("object")
mses
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LassoLars

data_p3 = pd.read_csv('Bias_correction_ucl.csv')
data_p3 = data_p3.dropna() 
data_p3 = data_p3.drop("station",axis=1)
data_p3 = data_p3.drop("Date",axis=1)
data_p3 = data_p3.drop("Next_Tmin",axis=1)

X_p3 = data_p3.sample(500)
Y_p3 = X_p3['Next_Tmax']

X_p3 = X_p3.drop('Next_Tmax',axis=1)

X_p3_train, X_p3_test, Y_p3_train, Y_p3_test = train_test_split(X_p3, Y_p3, test_size = 0.8)

Y_p3_train.describe()

LR = LinearRegression()
LR.fit(X_p3_train, Y_p3_train)

RT = DecisionTreeRegressor(random_state=0)
RT.fit(X_p3_train, Y_p3_train)

RFR = RandomForestRegressor(max_depth=2, random_state=0)
RFR.fit(X_p3_train, Y_p3_train)

SVR = SVR(C=1.0, epsilon=0.2)
SVR.fit(X_p3_train, Y_p3_train)

LL = LassoLars(alpha=0.01, normalize=False)
LL.fit(X_p3_train, Y_p3_train)

Y_lr_prediction = LR.predict(X_p3_test)
Y_rt_prediction = RT.predict(X_p3_test)
Y_rfr_prediction = RFR.predict(X_p3_test)
Y_svr_prediction = SVR.predict(X_p3_test)
Y_ll_prediction = LL.predict(X_p3_test)

Y_lr_prediction_tr = LR.predict(X_p3_train)
Y_rt_prediction_tr = RT.predict(X_p3_train)
Y_rfr_prediction_tr = RFR.predict(X_p3_train)
Y_svr_prediction_tr = SVR.predict(X_p3_train)
Y_ll_prediction_tr = LL.predict(X_p3_train)

rmse_lr_test = np.sqrt(mean_squared_error(Y_p3_test, Y_lr_prediction))
rmse_rt_test = np.sqrt(mean_squared_error(Y_p3_test, Y_rt_prediction))
rmse_rfr_test = np.sqrt(mean_squared_error(Y_p3_test, Y_rfr_prediction))
rmse_svr_test = np.sqrt(mean_squared_error(Y_p3_test, Y_svr_prediction))
rmse_ll_test = np.sqrt(mean_squared_error(Y_p3_test, Y_ll_prediction))

rmse_lr_tr = np.sqrt(mean_squared_error(Y_p3_train, Y_lr_prediction_tr))
rmse_rt_tr = np.sqrt(mean_squared_error(Y_p3_train, Y_rt_prediction_tr))
rmse_rfr_tr = np.sqrt(mean_squared_error(Y_p3_train, Y_rfr_prediction_tr))
rmse_svr_tr = np.sqrt(mean_squared_error(Y_p3_train, Y_svr_prediction_tr))
rmse_ll_tr = np.sqrt(mean_squared_error(Y_p3_train, Y_ll_prediction_tr))

rmse_mean = np.sqrt(mean_squared_error(Y_p3, [np.mean(Y_p3)] * 500))

cols = ['metric','linear regression','regression tree','random forest','support vector machine', 'lasso lars', 'mean']

rmses_test = np.column_stack([['rmse_test'] ,rmse_lr_test, rmse_rt_test, rmse_rfr_test, rmse_svr_test, rmse_ll_test, rmse_mean])
rmses_tr = np.column_stack([['rmse_tr'] ,rmse_lr_tr, rmse_rt_tr, rmse_rfr_tr, rmse_svr_tr, rmse_ll_tr,rmse_mean])

df_test = pd.DataFrame(data=rmses_test, columns=cols, dtype="object")
df_tr = pd.DataFrame(data=rmses_tr, columns=cols, dtype="object")

df3 = pd.concat([df_test, df_tr], ignore_index = True)

df3
```

On the training set RandomForestRegressor produces zero RMSE. If I am right, it is also the most complex model as the difference between rmse_tr and rmse_test is the largest. It has no error on the training set but produces some error on the test set. The best performing model seems to be good old Linear Regression. Results of the LassoLars is very similar because they actually almost the same algorithm (with alpha = 1 they would actually be the same). All regressors performed better than the dummy value mean.

```python
mse_comb = []

kf = KFold(n_splits=10)

rmse_lr_kfold = []
rmse_rt_kfold = []
rmse_rfr_kfold = []
rmse_svr_kfold = []
rmse_ll_kfold = []

for test, train in kf.split(list(zip(X_p3_train.iloc, Y_p3_train.iloc))):
        
    X_p3_kfold_train = [X_p3_train.iloc[i] for i in train]
    Y_p3_kfold_train = [Y_p3_train.iloc[i] for i in train]
        
    X_p3_kfold_test = [X_p3_train.iloc[i] for i in test]
    Y_p3_kfold_test = [Y_p3_train.iloc[i] for i in test]
                                    
    model = LinearRegression()
    model.fit(x_poly_comb_tr, y_comb_tr)
            
    y_poly_comb_pred = model.predict(x_poly_comb_test)
                
    LR = LinearRegression()
    LR.fit(X_p3_kfold_train, Y_p3_kfold_train)

    RT = DecisionTreeRegressor(random_state=0)
    RT.fit(X_p3_kfold_train, Y_p3_kfold_train)

    RFR = RandomForestRegressor(max_depth=2, random_state=0)
    RFR.fit(X_p3_kfold_train, Y_p3_kfold_train)

    # SVR = SVR(C=1.0, epsilon=0.2)
    SVR.fit(X_p3_kfold_train, Y_p3_kfold_train)

    LL = LassoLars(alpha=0.01, normalize=False)
    LL.fit(X_p3_kfold_train, Y_p3_kfold_train)

    Y_lr_prediction_kfold = LR.predict(X_p3_kfold_test)
    Y_rt_prediction_kfold = RT.predict(X_p3_kfold_test)
    Y_rfr_prediction_kfold = RFR.predict(X_p3_kfold_test)
    Y_svr_prediction_kfold = SVR.predict(X_p3_kfold_test)
    Y_ll_prediction_kfold = LL.predict(X_p3_kfold_test)
    
    rmse_lr_kfold.append(np.sqrt(mean_squared_error(Y_p3_kfold_test, Y_lr_prediction_kfold)))
    rmse_rt_kfold.append(np.sqrt(mean_squared_error(Y_p3_kfold_test, Y_rt_prediction_kfold)))
    rmse_rfr_kfold.append(np.sqrt(mean_squared_error(Y_p3_kfold_test, Y_rfr_prediction_kfold)))
    rmse_svr_kfold.append(np.sqrt(mean_squared_error(Y_p3_kfold_test, Y_svr_prediction_kfold)))
    rmse_ll_kfold.append(np.sqrt(mean_squared_error(Y_p3_kfold_test, Y_ll_prediction_kfold)))

rmse_lr_kfold = np.mean(rmse_lr_kfold)
rmse_rt_kfold = np.mean(rmse_rt_kfold)
rmse_rfr_kfold = np.mean(rmse_rfr_kfold)
rmse_svr_kfold = np.mean(rmse_svr_kfold)
rmse_ll_kfold = np.mean(rmse_ll_kfold)

rmses_kfold = np.column_stack([['rmse_kfold'], rmse_lr_kfold, rmse_rt_kfold, rmse_rfr_kfold, rmse_svr_tr, rmse_ll_kfold, rmse_mean])

df_kfold = pd.DataFrame(data=rmses_kfold, columns=cols, dtype="object")

df3 = pd.concat([df_test, df_tr, df_kfold], ignore_index = True)

df3
```
